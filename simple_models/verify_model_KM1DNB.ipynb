```python
import matplotlib.pyplot as plt
from single_solution_KM1D2B import DualBubble
from single_solution_KM1DNBC import MultiBubble
from single_solution_KM1DNBC import P0, CL
import numpy as np

plt.rcParams.update({
    "text.usetex": False,
    "font.size"  : 16})

LR = CL / (20 * 1000)
```

## Compare DualBubble and MultiBubble Results

```python
# Figure 2 (a)
R0 = [7, 5]
PA = [1.2 * P0 * 1e-5]
dual_bubble = DualBubble(R0, [20], PA, k=1, AC_FIELD="CONST")
dual_bubble.x0[1] = 200 *1e-6 / LR
dual_bubble.T     = 10

multi_bubble = MultiBubble(R0, [20], PA, [0], k=1, AC_FIELD="CONST")
multi_bubble.x0[1] = dual_bubble.x0[1]
multi_bubble.T     = dual_bubble.T

t2, _, _, x2, _ = dual_bubble.integrate()
tN, _, _, xN, _ = multi_bubble.integrate()

plt.figure(figsize=(10, 6))
plt.plot(t2, x2[0] * LR * 1e6, "r-", label="Bubble-1")
plt.plot(t2, x2[1] * LR * 1e6, "b-", label="Bubble-2")
plt.plot(tN, xN[0] * LR * 1e6, "g--", label="Bubble-1")
plt.plot(tN, xN[1] * LR * 1e6, "m--", label="Bubble-2")
plt.xlabel(r'$\tau$')
plt.ylabel(r'$z\, \mu m$')
plt.legend()
plt.grid('both')
plt.show()
```

```python
# Figure 4 (a)
R0 = [6, 5]
PA = [-1.2 * P0 * 1e-5]
dual_bubble = DualBubble(R0, [20], PA, k=1, AC_FIELD="CONST")
dual_bubble.x0[1] = 300 *1e-6 / LR
dual_bubble.T = 8.0
multi_bubble = MultiBubble(R0, [20], PA, [0], k=1, AC_FIELD="CONST")
multi_bubble.x0[1] = dual_bubble.x0[1]
multi_bubble.T = dual_bubble.T

t2, r2, _, x2, _ = dual_bubble.integrate()
tN, rN, _, xN, _ = dual_bubble.integrate()

plt.figure(figsize=(10, 6))
plt.plot(t2, (x2[1]-x2[0]) * LR * 1e6, "r-", label="D(t)")
plt.plot(t2, (r2[0]*R0[0] + r2[1]*R0[1]), "b-", label="$R_0(t)+R_1(t)$")
plt.plot(tN, (xN[1]-xN[0]) * LR * 1e6, "g--", label="D(t)")
plt.plot(tN, (rN[0]*R0[0] + rN[1]*R0[1]), "m--", label="$R_0(t)+R_1(t)$")
plt.xlabel(r'$\tau$')
plt.ylabel(r'$z\, \mu m$')
plt.legend()
plt.grid('both')
plt.show()
```

# Multi-Bubble Scenarios

```python
# 3 bubble system symmetric arrangement 
R0 = [7, 5, 7]
PA = [1.2 * P0 * 1e-5]

multi_bubble = MultiBubble(R0, [20], PA, [0], k=1, AC_FIELD="CONST")
multi_bubble.x0[1] = 500 *1e-6 / LR
multi_bubble.x0[2] = 1000 *1e-6 / LR
multi_bubble.T     = 20

t, _, _, x, _ = multi_bubble.integrate()

plt.figure(figsize=(10, 6))
plt.plot(t, x[0] * LR * 1e6, "r-", label="Bubble-1")
plt.plot(t, x[1] * LR * 1e6, "b-", label="Bubble-2")
plt.plot(t, x[2] * LR * 1e6, "g-", label="Bubble-3")
plt.xlabel(r'$\tau$')
plt.ylabel(r'$z\, \mu m$')
plt.legend()
plt.grid('both')
plt.show()
```

# Chain of Bubbles

```python
# 8 bubble system
R0 = [10] * 8
PA = [1.2 * P0 * 1e-5]

multi_bubble = MultiBubble(R0, [20], PA, [0], k=1, AC_FIELD="CONST")
multi_bubble.x0 = np.linspace(0, 5000 * 1e-6 / LR, len(R0), endpoint=True)
multi_bubble.T  = 20

t, _, _, x, _ = multi_bubble.integrate()

plt.figure(figsize=(10, 6))
for i in range(len(R0)):
    plt.plot(t, x[i] * LR * 1e6, "-", label=f"Bubble-{i}")

plt.xlabel(r'$\tau$')
plt.ylabel(r'$z\, \mu m$')
plt.legend()
plt.grid('both')
plt.show()
```
