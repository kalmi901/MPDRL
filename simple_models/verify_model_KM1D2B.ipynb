This notebook intends to verify the implementation of the model `KM1D2B` by comparing the dimensionless time-radius curves at various parameters.

From the paper `A.A. Doinikov, TTranslational motion of two interacting bubbles in a strong acoustic field, Phys. Rev. E, 64, (2001), p. 026301`, multiple figures are recalculated to verify the present implementation. 

```python
import matplotlib.pyplot as plt
from single_solution_KM1D2B import DualBubble as KMBubbles
from single_solution_KM1D2B import P0, CL

plt.rcParams.update({
    "text.usetex": False,
    "font.size"  : 16})
```

```python
LR = CL / (20 * 1000)       # Reference Wave-Length
```

```python
# Figure 2 (a)
R0 = [7, 5]
PA = [1.2 * P0 * 1e-5]
km = KMBubbles(R0, [20], PA, k=1, AC_FIELD="CONST")
km.x0[1] = 200 *1e-6 / LR
km.T = 100

t, _, _, x, _ = km.integrate()

plt.figure(figsize=(10, 6))
plt.plot(t, x[0] * LR * 1e6, "r-", label="Bubble-1")
plt.plot(t, x[1] * LR * 1e6, "b-", label="Bubble-2")
plt.xlabel(r'$\tau$')
plt.ylabel(r'$z\, \mu m$')
plt.legend()
plt.grid('both')
plt.show()
```
